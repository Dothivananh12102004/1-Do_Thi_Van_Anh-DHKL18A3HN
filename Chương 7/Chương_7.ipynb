```python
import tkinter as tk

def calc_sum():
    try:
        n = int(entry.get())
        total = sum(range(1, n + 1))
        result_label.config(text=f"1 + 2 + ... + {n} = {total}")
    except:
        result_label.config(text="Vui lòng nhập số nguyên!")

root = tk.Tk()
root.title("Tính tổng 1..N")

tk.Label(root, text="Nhập N:").pack()

entry = tk.Entry(root, width=20)
entry.pack(pady=5)

btn = tk.Button(root, text="Tính", command=calc_sum)
btn.pack(pady=10)

result_label = tk.Label(root, text="", font=("Arial", 14))
result_label.pack()

root.mainloop()
```

```python
import tkinter as tk

def list_divisors():
    try:
        n = int(entry.get())
        divisors = [str(i) for i in range(1, n+1) if n % i == 0]
        result.config(text="The divisors of N:   " + "   ".join(divisors))
    except:
        result.config(text="Please enter an integer!")

root = tk.Tk()
root.title("Divisors of N")

tk.Label(root, text="Enter the value of N").pack()
entry = tk.Entry(root)
entry.pack()

result = tk.Label(root, text="", fg="blue")
result.pack(pady=10)

tk.Button(root, text="Validate", command=list_divisors).pack()

root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk
import math

def calculate():
    try:
        m = int(entry_m.get())
        n = int(entry_n.get())
        func = combo.get()

        if func == "GCD":
            value = math.gcd(m, n)
            result.config(text=f"GCD(m, n) = {value}")
        else:
            value = abs(m*n) // math.gcd(m, n)
            result.config(text=f"LCM(m, n) = {value}")
    except:
        result.config(text="Invalid input!")

root = tk.Tk()
root.title("GCD & LCM Calculator")

tk.Label(root, text="Enter value of m:").pack()
entry_m = tk.Entry(root)
entry_m.pack()

tk.Label(root, text="Enter value of n:").pack()
entry_n = tk.Entry(root)
entry_n.pack()

tk.Label(root, text="Choose function:").pack()
combo = ttk.Combobox(root, values=["GCD", "LCM"])
combo.current(0)
combo.pack()

result = tk.Label(root, text="", fg="blue")
result.pack(pady=10)

tk.Button(root, text="Validate", command=calculate).pack()

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

def show_info():
    name = entry_name.get()
    try:
        age = int(entry_age.get())
    except:
        messagebox.showerror("Error", "Tuổi phải là số!")
        return
    
    msg = f"Xin chào {name}!\n"
    if age >= 18:
        msg += "Bạn đã trưởng thành! Tuổi đã nhập >= 18"
    else:
        msg += "Bạn còn nhỏ tuổi! Tuổi đã nhập < 18"

    messagebox.showinfo("Thông báo", msg)

root = tk.Tk()
root.title("Thông tin người dùng")

tk.Label(root, text="Nhập tên:").pack()
entry_name = tk.Entry(root)
entry_name.pack()

tk.Label(root, text="Nhập tuổi:").pack()
entry_age = tk.Entry(root)
entry_age.pack()

tk.Button(root, text="Validate", command=show_info).pack()

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

animals = [
    "Thân", "Dậu", "Tuất", "Hợi", "Tý", "Sửu",
    "Dần", "Mão", "Thìn", "Tỵ", "Ngọ", "Mùi"
]

def convert_year():
    try:
        y = int(entry.get())
        zodiac = animals[y % 12]
        messagebox.showinfo("Năm âm lịch", f"Năm {y} là năm con {zodiac}")
    except:
        messagebox.showerror("Lỗi", "Hãy nhập số năm hợp lệ!")

root = tk.Tk()
root.title("Âm lịch")

tk.Label(root, text="Nhập năm dương lịch:").pack()
entry = tk.Entry(root)
entry.pack()

tk.Button(root, text="Chuyển đổi", command=convert_year).pack()

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

def calc_bmi():
    try:
        weight = float(entry_weight.get())
        height = float(entry_height.get())
        bmi = weight / (height * height)

        if bmi < 18.5:
            status = "Gầy"
        elif bmi < 25:
            status = "Bình thường"
        elif bmi < 30:
            status = "Thừa cân"
        else:
            status = "Béo phì"

        result.config(text=f"BMI: {bmi:.2f}\nKết luận: {status}")

    except:
        messagebox.showerror("Lỗi", "Dữ liệu không hợp lệ!")

root = tk.Tk()
root.title("BMI Calculator")

tk.Label(root, text="Cân nặng (kg):").pack()
entry_weight = tk.Entry(root)
entry_weight.pack()

tk.Label(root, text="Chiều cao (m):").pack()
entry_height = tk.Entry(root)
entry_height.pack()

tk.Button(root, text="Tính BMI", command=calc_bmi).pack()

result = tk.Label(root, text="", fg="blue")
result.pack()

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

def show_message(name):
    messagebox.showinfo("Lập trình hướng đối tượng", 
                        f"Viết phương trình giải phương trình bậc hai.")

root = tk.Tk()
root.title("Hệ thống bài tập nâng cao Python")

menubar = tk.Menu(root)

ch1 = tk.Menu(menubar, tearoff=0)
for i in range(1, 10):
    ch1.add_command(label=f"1.{i}", command=lambda x=i: show_message(f"1.{x}"))
ch1.add_separator()
ch1.add_command(label="Exit", command=root.quit)

menubar.add_cascade(label="Chương 1", menu=ch1)
menubar.add_cascade(label="Chương 2", menu=tk.Menu(menubar))
menubar.add_cascade(label="Chương 3", menu=tk.Menu(menubar))
menubar.add_cascade(label="Chương 4", menu=tk.Menu(menubar))

root.config(menu=menubar)
root.mainloop()
```

```python
import tkinter as tk

def calc():
    try:
        km = float(entry_km.get())
        wait = int(entry_wait.get())
        car_type = var.get()

        # Biểu giá
        if car_type == 4:
            price_open = 12000
            price_1 = 15300
            price_2 = 12100
        else:
            price_open = 12000
            price_1 = 16100
            price_2 = 13800

        total = 0

        # Tính tiền km
        if km <= 0.8:
            total = price_open
        else:
            total = price_open
            remaining = km - 0.8

            if remaining <= 29.2:
                total += remaining * price_1
            else:
                total += 29.2 * price_1
                total += (remaining - 29.2) * price_2

        # Tính tiền chờ
        if wait > 5:
            total += (wait - 5) * 750

        result.config(text=f"Tổng tiền: {total:,.0f} đ")

    except:
        result.config(text="Lỗi dữ liệu đầu vào!")

root = tk.Tk()
root.title("Taxi Fare Calculator")

tk.Label(root, text="Nhập số km đã đi:").pack()
entry_km = tk.Entry(root)
entry_km.pack()

tk.Label(root, text="Thời gian chờ (phút):").pack()
entry_wait = tk.Entry(root)
entry_wait.pack()

tk.Label(root, text="Chọn loại xe:").pack()
var = tk.IntVar(value=4)
tk.Radiobutton(root, text="Xe 4 chỗ", variable=var, value=4).pack()
tk.Radiobutton(root, text="Xe 7 chỗ", variable=var, value=7).pack()

tk.Button(root, text="Tính tiền", command=calc).pack(pady=5)

result = tk.Label(root, text="", fg="blue")
result.pack()

root.mainloop()
```
