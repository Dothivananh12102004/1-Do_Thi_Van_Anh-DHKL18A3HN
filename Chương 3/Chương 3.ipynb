Bài 3.8:

```python
import numpy as np

# Đọc dữ liệu
heights = np.loadtxt("heights.txt")
positions = np.loadtxt("positions.txt", dtype=str)

np_positions = np.array(positions)
np_heights = np.array(heights)

# GK
gk_heights = np_heights[np_positions == 'GK']
print("Chiều cao trung bình GK:", gk_heights.mean())

# Các vị trí khác GK
other_heights = np_heights[np_positions != 'GK']
print("Chiều cao trung bình các vị trí khác:", other_heights.mean())

# Tạo structured array
players = np.zeros(len(np_heights), dtype=[('position','U5'), ('height','float')])
players['position'] = np_positions
players['height'] = np_heights

# Sắp xếp theo height
players_sorted = np.sort(players, order='height')

print("Vị trí cao nhất:", players_sorted[-1])
print("Vị trí thấp nhất:", players_sorted[0])
```

Bài 3.11:

```python
import pandas as pd

# a) stocks1
stocks1 = pd.read_csv("stocks1.csv")
print(stocks1.head())
print(stocks1.tail())
print(stocks1.dtypes)
print(stocks1.info())

# b) stocks2
stocks2 = pd.read_csv("stocks2.csv")
print(stocks2.head())
print(stocks2.tail())
print(stocks2.dtypes)
print(stocks2.info())

# c) companies
companies = pd.read_csv("companies.csv")
print(companies.head())
print(companies.dtypes)
print(companies.info())

# 2. Kiểm tra Null
print(stocks1.isnull().sum())

# Thay thế Null
stocks1['high'].fillna(stocks1['high'].max(), inplace=True)
stocks1['low'].fillna(stocks1['low'].min(), inplace=True)

# 3. Gộp stocks1 và stocks2 theo dòng
import pandas as pd

stocks = pd.concat([stocks1, stocks2], axis=0, ignore_index=True)

# Xem 15 dòng cuối
stocks.tail(15)

# 4. Gộp stocks và companies
stocks_companies = pd.merge(
    stocks,
    companies,
    on="company_id",
    how="left"
)

# Xem 5 dòng đầu
stocks_companies.head()

#5. Giá (open, high, low, close) TB & volume TB mỗi công ty
avg_price_volume = stocks_companies.groupby("company_name").agg(
    open_avg=("open", "mean"),
    high_avg=("high", "mean"),
    low_avg=("low", "mean"),
    close_avg=("close", "mean"),
    volume_avg=("volume", "mean")
)

avg_price_volume

# 6. Giá đóng cửa trung bình, lớn nhất, nhỏ nhất
close_stats = stocks_companies.groupby("company_name")["close"].agg(
    close_mean="mean",
    close_max="max",
    close_min="min"
)

close_stats

# 7. Tạo cột parsed_time (Datetime)
stocks_companies["parsed_time"] = pd.to_datetime(stocks_companies["time"])

# Kiểu dữ liệu
stocks_companies["parsed_time"].dtype

# Hiển thị 5 dòng đầu
stocks_companies.head()

# 8. Thêm cột result (up / down)
stocks_companies["result"] = stocks_companies.apply(
    lambda x: "up" if x["close"] > x["open"] else "down",
    axis=1
)

stocks_companies[["open", "close", "result"]].head()
```

```python
# Đọc dữ liệu:
movies = pd.read_csv("movies.csv")
ratings = pd.read_csv("ratings.csv")
tags = pd.read_csv("tags.csv")

# Kiểm tra & xử lý dữ liệu thiếu:
movies.isnull().sum()
ratings.isnull().sum()
tags.isnull().sum()

# Ví dụ xử lý
tags = tags.dropna()

# Gộp DataFrame:
movie_ratings = pd.merge(
    ratings,
    movies,
    on="movieId",
    how="left"
)

# Parsing Timestamp:
movie_ratings["rating_time"] = pd.to_datetime(
    movie_ratings["timestamp"], unit="s"
)

movie_ratings.dtypes

# Lọc dữ liệu theo yêu cầu:
# Phim có rating >= 4
high_rating_movies = movie_ratings[movie_ratings["rating"] >= 4]

# Thống kê dữ liệu:
stats = movie_ratings.groupby("title").agg(
    avg_rating=("rating", "mean"),
    rating_count=("rating", "count")
).sort_values("rating_count", ascending=False)

stats.head(10)
```
