```python
import pandas as pd
import numpy as np
# Đọc dữ liệu CSV:
# Đọc stocks1.csv:
stocks1 = pd.read_csv('stocks1.csv')

# 5 dòng đầu và cuối:
print(stocks1.head())
print(stocks1.tail())

# Kiểu dữ liệu
print(stocks1.dtypes)

# Thông tin tổng quát
print(stocks1.info())

# Đọc stocks2.csv:
stocks2 = pd.read_csv('stocks2.csv')

print(stocks2.head())
print(stocks2.tail())
print(stocks2.dtypes)
print(stocks2.info())

# Đọc companies.csv:
companies = pd.read_csv('companies.csv')

print(companies.head())
print(companies.dtypes)
print(companies.info())

# Xử lý giá trị Null trong stocks1:
# Kiểm tra Null:
print(stocks1.isnull().sum())

# Thay thế theo yêu cầu:
stocks1['high'] = stocks1.groupby('symbol')['high'] \
                          .transform(lambda x: x.fillna(x.max()))

stocks1['low'] = stocks1.groupby('symbol')['low'] \
                         .transform(lambda x: x.fillna(x.min()))

# Gộp stocks1 và stocks2 theo dòng:
stocks = pd.concat([stocks1, stocks2], ignore_index=True)

# Xem 15 dòng cuối
print(stocks.tail(15))

# Gộp stocks với companies:

# Gộp theo cột symbol:

stocks_companies = pd.merge(
    stocks,
    companies,
    left_on='symbol',
    right_on='name',
    how='left'
)

# 5 dòng đầu:
print(stocks_companies.head())

# Giá trung bình & volume trung bình của mỗi công ty:
avg_prices = stocks_companies.groupby('symbol')[[
    'open', 'high', 'low', 'close', 'volume'
]].mean()
print(avg_prices)

# Giá đóng cửa(close) trung bình, lớn nhất, nhỏ nhất:
close_stats = stocks_companies.groupby('symbol')['close'] \
    .agg(['mean', 'max', 'min'])
print(close_stats)

# Tạo cột parsed_time(DateTime):
stocks_companies['parsed_time'] = pd.to_datetime(
    stocks_companies['date'],
    format='mixed'
)

# Kiểu dữ liệu
print(stocks_companies['parsed_time'].dtype)

# 5 dòng đầu
print(stocks_companies.head())

# Thêm cột result:
stocks_companies['result'] = np.where(
    stocks_companies['close'] > stocks_companies['open'],
    'up',
    'down'
)
print(stocks_companies[['open', 'close', 'result']].head())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Đọc dữ liệu & Data Structures:
# Đọc các file CSV:
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
tags = pd.read_csv('tags.csv')
links = pd.read_csv('links.csv')

# Xem 5 dòng đầu:
print(movies.head())
print(ratings.head())
print(tags.head())
print(links.head())

# Kiểu dữ liệu & kích thước:
print(movies.dtypes)
print(movies.shape)
print(ratings.dtypes)
print(ratings.shape)

# Kiểm tra giá trị thiếu (Null)
print(movies.isnull().sum())
print(ratings.isnull().sum())
print(tags.isnull().sum())
print(links.isnull().sum())
# GỘP DATAFRAME:
# Gộp ratings với movies:
movie_ratings = pd.merge(
    ratings,
    movies,
    on='movieId',
    how='inner'
)
print(movie_ratings.head())

# LỌC DỮ LIỆU THEO YÊU CẦU:
#2.1 Lọc rating hợp lệ (0.5 → 5.0):
movie_ratings = movie_ratings[
    (movie_ratings['rating'] >= 0.5) &
    (movie_ratings['rating'] <= 5)
]

#2.2 Lọc phim thuộc thể loại Action:
action_movies = movie_ratings[
    movie_ratings['genres'].str.contains('Action', na=False)
]
print(action_movies.head())

#2.3 Lọc phim có trên 100 lượt đánh giá:
rating_count = movie_ratings.groupby('title').size()
popular_movies = rating_count[rating_count > 100].index
filtered_movies = movie_ratings[
    movie_ratings['title'].isin(popular_movies)
]
# THỐNG KÊ DỮ LIỆU:
# 3.1 Rating trung bình mỗi phim:
avg_rating = movie_ratings.groupby('title')['rating'].mean()
print(avg_rating.sort_values(ascending=False).head())

# 3.2 Top 10 phim được đánh giá nhiều nhất:
top10_movies = movie_ratings['title'].value_counts().head(10)
print(top10_movies)

# 3.3 Rating trung bình theo thể loại:
genre_rating = (
    movie_ratings
    .assign(genres=movie_ratings['genres'].str.split('|'))
    .explode('genres')
    .groupby('genres')['rating']
    .mean()
)
print(genre_rating.sort_values(ascending=False))

# PARSING TIMESTAMPS:
# Chuyển timestamp (UNIX → DateTime):
movie_ratings['timestamp'] = pd.to_datetime(
    movie_ratings['timestamp'],
    unit='s'
)
print(movie_ratings['timestamp'].dtype)
print(movie_ratings.head())

# Tách năm rating:
movie_ratings['year'] = movie_ratings['timestamp'].dt.year

#️ TRỰC QUAN HÓA DỮ LIỆU:
# 5.1 Số lượng rating theo năm:
ratings_per_year = movie_ratings['year'].value_counts().sort_index()
plt.figure()
plt.plot(ratings_per_year.index, ratings_per_year.values)
plt.xlabel('Year')
plt.ylabel('Number of Ratings')
plt.title('Number of Ratings per Year')
plt.show()
# 5.2 Phân phối rating:
plt.figure()
plt.hist(movie_ratings['rating'], bins=10)
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Ratings')
plt.show()

```
